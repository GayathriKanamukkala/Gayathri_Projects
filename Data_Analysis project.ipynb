```python
def group_and_calculate_percentage(df, column_name, threshold=50000):
    # Melt the DataFrame to have one row per sale entry
    melted_df = df.melt(id_vars=['Dress_ID', column_name], var_name='Date', value_name='Sales')

    # Function to extract month and determine season
    def extract_month(date_str):
        try:
            return int(date_str.split('-')[1])
        except (IndexError, ValueError):
            return None

    melted_df['Month'] = melted_df['Date'].apply(extract_month)

    # Remove rows where month extraction failed
    melted_df = melted_df.dropna(subset=['Month'])

    def get_season(month):
        if month in [6, 7, 8]:
            return 'Summer'
        elif month in [9, 10, 11]:
            return 'Autumn'
        elif month in [12, 1, 2]:
            return 'Winter'
        elif month in [3, 4, 5]:
            return 'Spring'

    melted_df['Season'] = melted_df['Month'].apply(get_season)

    # Calculate the total sales for each category across all seasons
    total_sales_per_category = melted_df.groupby(column_name)['Sales'].sum().reset_index()

    # Identify categories with total sales less than threshold
    categories_to_group = total_sales_per_category[total_sales_per_category['Sales'] < threshold][column_name].tolist()

    # Replace those categories with "Others"
    melted_df[column_name] = melted_df[column_name].apply(lambda x: 'Others' if x in categories_to_group else x)

    # Group by the specified column and Season, and sum the sales
    grouped_sales = melted_df.groupby([column_name, 'Season'])['Sales'].sum().reset_index()

    # Calculate the total sales for each category
    category_totals = grouped_sales.groupby(column_name)['Sales'].sum().reset_index()

    # Calculate the percentage of each category
    grouped_sales = grouped_sales.merge(category_totals, on=column_name, suffixes=('', '_Total'))
    grouped_sales['Percentage'] = (grouped_sales['Sales'] / grouped_sales['Sales_Total']) * 100

    # Return only the grouped columns and percentage
    return grouped_sales[[column_name, 'Season', 'Percentage']]
```

```python
# Group "Style" categories into "Others" which have less than 50000 sales across all the seasons.
data = {
    'Style': ['Style1', 'Style2', 'Style3', 'Style4', 'Style5'],
    'Season1': [10000, 20000, 3000, 15000, 40000],
    'Season2': [20000, 30000, 5000, 10000, 5000],
    'Season3': [10000, 10000, 2000, 5000, 5000]
}

df = pd.DataFrame(data)

# Calculate total sales across all seasons for each style
df['TotalSales'] = df[['Season1', 'Season2', 'Season3']].sum(axis=1)

# Identify styles with less than 50,000 total sales
df_low_sales = df[df['TotalSales'] < 50000]

# Sum the sales of these low-selling styles to group into "Others"
others_sales = df_low_sales[['Season1', 'Season2', 'Season3']].sum().to_frame().T
others_sales['Style'] = 'Others'
others_sales['TotalSales'] = others_sales[['Season1', 'Season2', 'Season3']].sum(axis=1)

# Filter out the low-selling styles from the original DataFrame
df_high_sales = df[df['TotalSales'] >= 50000]

# Append the "Others" row to the high-selling styles DataFrame
df_result = pd.concat([df_high_sales, others_sales], ignore_index=True)

# Drop the 'TotalSales' column if not needed in the final result
df_result = df_result.drop(columns=['TotalSales'])

print(df_result)
```
